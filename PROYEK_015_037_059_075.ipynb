$$
BST\ yang\ telah\ disimpan\ sebelumnya\
b:\ kata\ (dalam\ Bahasa\ Indonesia)\
$$
Keluaran $(output)$: \
$$
BST\ yang\ menyisipkan\ kata\ b\ dalam\ BST\ sebelumnya\
$$
\
\
$Catatan:$ \
Jika Anda mengalami kesulitan dalam menggambarkan BST, Anda dapat

Jawaban Soal A

```python
# Program Konstruksi BST

# Input
n = int(input("Masukkan jumlah kata: "))
words = []
for i in range(n):
    word = input("Masukkan kata: ")
    words.append(word)

# Konstruksi BST
bst = []
for i in range(n):
    word = words[i]
    # Akar (Pada awalnya tidak punya anak)
    node = [word, None, None]
    if i == 0:
        bst.append(node)
    else:
        current = bst[0]
        while True:
            # Mengurutkan secara leksikografis (alfabetis)
            if word < current[0]:
                if current[1] is None:
                    current[1] = node
                    break
                else:
                    current = current[1]
            else:
                if current[2] is None:
                    current[2] = node
                    break
                else:
                    current = current[2]

# Output BST
sisi = []
stack = [bst[0]]
while stack:
    current = stack.pop()
    if current[1] is not None:
        sisi.append((current[0], current[1][0]))
        stack.append(current[1])
    if current[2] is not None:
        sisi.append((current[0], current[2][0]))
        stack.append(current[2])

output = ", ".join([f"({x[0]},{x[1]})" for x in sisi])
print("BST hasil konstruksi:")
print(output)
```

Jawaban Soal B

```python
# Program Sisipkan kata dalam BST
b = input("Masukkan kata yang ingin disisipkan: ")

# Titik dalam
node = [b, None, None]
current = bst[0]
while True:
    # Mengurutkan kata baru secara leksikografis
    if b < current[0]:
        if current[1] is None:
            current[1] = node
            break
        else:
            current = current[1]
    else:
        if current[2] is None:
            current[2] = node
            break
        else:
            current = current[2]

# Output BST setelah penyisipan
sisi = []
stack = [bst[0]]  # Urut secara vertikal
while stack:
    current = stack.pop()
    if current[1] is not None:
        sisi.append((current[0], current[1][0]))
        stack.append(current[1])
    if current[2] is not None:
        sisi.append((current[0], current[2][0]))
        stack.append(current[2])

output = ", ".join([f"({x[0]},{x[1]})" for x in sisi])
print("BST setelah penyisipan kata:")
print(output)
```
